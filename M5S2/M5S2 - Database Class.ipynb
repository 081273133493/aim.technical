```python
import sqlite3
import pandas as pd
from IPython.display import display
```

```python
class DB:
    def __init__(self, db_name):
        self.db_name = db_name
        self.conn = sqlite3.connect(db_name)
        self.curr = self.conn.cursor()
        
    def run_query(self, query, show=True):
        self.curr.execute(query)
        result = self.curr.fetchall()
        if show:
            columns = [column[0] for column in self.curr.description]
            df = pd.DataFrame(result, columns=columns)
            display(df)
        else:    
            return result
        
    def show_tables(self):
        query = "SELECT name FROM sqlite_master WHERE type='table';"
        columns = self.curr.execute(query)
        columns = [column[0] for column in columns]
        for column in columns:
            print(column)
        
    def close(self):
        self.conn.close()

    def describe_table(self, table_name):
        table_info = self.run_query(f"PRAGMA table_info({table_name})", show=False)
        columns = [column[0] for column in self.curr.description]
        df = pd.DataFrame(table_info, columns=columns).set_index('name')
        df.drop(labels='cid', axis=1, inplace=True)
        display(df)
```

```python
data_penjualan = DB('data_penjualan.db')
```

```python
data_penjualan.show_tables()
```

```python
data_penjualan.describe_table('orders')
```
